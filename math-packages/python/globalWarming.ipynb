```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('stockholm_tmp.dat')
```

Задание:
1. Определить среднюю температуру по месяцам и построить график
2. Определить как в апреле менялась погода на протяжении всех лет, построить график 

```python
ms = np.unique(data[:, 1])
print(ms)
srs = []
for i in ms:
    x = data[np.where(data[:, 1] == i)]
    
    srs.append(np.sum(x[:, 3])/len(x))
print(srs)
```

```python
plt.hist(range(1, len(srs)+1), bins=len(srs), weights=srs, rwidth=0.9, color='teal') 

plt.xlabel("Месяц")
plt.ylabel("Средняя температура")
plt.show()
```

```python
ms = 4.
april_data = data[np.where(data[:, 1] == ms)]
years = np.unique(april_data[:, 0])
ans = []
for i in years:
    x = data[np.where((data[:, 0] == i))]
    ans.append(np.sum(x[:, 3])/len(x))

plt.figure(figsize=(20, 10))
plt.plot(years, ans, color = 'teal')
plt.xlabel("Год", size = 15)
plt.ylabel("Средняя тепература в апреле", size = 15)


plt.show()
```
